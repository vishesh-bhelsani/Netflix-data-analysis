```python
import pandas as pd
df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.fillna("unknown", inplace = True)
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt 
df["type"].value_counts().plot(kind = "bar", color = ["red", "black"])
plt.title("movies vs Tv shows on netflix")
plt.xlabel("Type")
plt.ylabel("count")
plt.show()
```

```python
df["country"].value_counts().head(10).plot(kind = "bar", color = "blue")
plt.title("top 10 companies on netflicx")
plt.xlabel("country")
plt.ylabel("number of titles")
plt.tight_layout()
plt.show()
```

```python
df["listed_in"].str.split(", ").explode().value_counts().head(10).plot(kind = "bar", color = "lightblue")
plt.title("top 10 genre on netflix")
plt.xlabel("Genre")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors = "coerce")
df["year_added"] = df["date_added"].dt.year
df["year_added"].value_counts().sort_index().plot(kind = "line", color = "lightgreen",marker = "o")
plt.title("content added to netflix each year")
plt.xlabel("year")
plt.ylabel("number of titles")
plt.tight_layout()
plt.show()
```

```python
df["rating"].value_counts().head(8).plot(kind = "bar", color= "grey")
plt.title("most common rating on netflix")
plt.xlabel("rating")
plt.ylabel("count")
plt.tight_layout()
plt.show()
```

#key findings
1. netflix have more movies than tv shows.
2. USA produce the most content on netflix
3. international movies and drama are the most popular genres.
4. netflix grew rapidly between 2015- 2019.
5. netflix focus mostly on adult content (TV-MA)
